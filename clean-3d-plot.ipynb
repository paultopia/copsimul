```python
import pandas as pd
import numpy as np
import matplotlib  
%matplotlib inline
import matplotlib.pyplot as plt
import plottyprint
from mpl_toolkits import mplot3d
df = pd.read_json("sorted_simul_dump.json")
```

```python
fig = plt.figure(figsize=(10, 10))
plt.tight_layout()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df["segregation"], df["bias"], df["inequality"], c='gray', marker='.')
ax.locator_params(nbins=3)
ax.set_xlabel('Segregation')
ax.set_ylabel('Bias Level')
ax.set_zlabel('Inequality', rotation="horizontal")
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
```

```python
fig2 = plt.figure(figsize=(10, 10))
plt.tight_layout()
ax2 = fig2.add_subplot(111, projection='3d')
ax2.scatter(df["segregation"], df["bias"], df["inequality"], c='gray', marker='.')
ax2.locator_params(nbins=3)
ax2.set_xlabel('Segregation', fontsize=15, labelpad=10, fontname='Lato')
ax2.set_ylabel('Bias Level', fontsize=15, labelpad=10, fontname='Lato')
ax2.set_zlabel('Inequality', fontsize=15, labelpad=10, fontname='Lato')
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.spines['left'].set_visible(False)
ax2.spines['bottom'].set_visible(False)
```

```python
fig3 = plt.figure(figsize=(12, 10))
plt.tight_layout()
ax3 = fig3.add_subplot(111, projection='3d')
ax3.set_title("Predictive Policing Bias Simulation" + '\n', fontsize=20, fontname='Lato')
ax3.scatter(df["segregation"], df["bias"], df["inequality"], c='gray', marker='.')
ax3.locator_params(nbins=3)
ax3.set_xlabel('Segregation', fontsize=15, labelpad=10, fontname='Lato', color = (0.15, 0.15, 0.15),)
ax3.set_ylabel('Bias Level', fontsize=15, labelpad=10, fontname='Lato', color = (0.15, 0.15, 0.15),)
ax3.set_zlabel('Inequality', fontsize=15, labelpad=10, fontname='Lato', color = (0.15, 0.15, 0.15),)
ax3.spines['top'].set_visible(False)
ax3.spines['right'].set_visible(False)
ax3.spines['left'].set_visible(False)
ax3.spines['bottom'].set_visible(False)
ax3.xaxis.pane.fill = False
ax3.yaxis.pane.fill = False
ax3.zaxis.pane.fill = False
ax3.xaxis.pane.set_edgecolor('w')
ax3.yaxis.pane.set_edgecolor('w')
ax3.zaxis.pane.set_edgecolor('w')
```
